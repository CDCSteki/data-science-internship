# INITIAL SETUP
Read data set

```python
import pandas as pd
import os
import numpy as np
import seaborn as sns                      
import matplotlib.pyplot as plt
from datetime import datetime as dt
            
%matplotlib inline     
sns.set(color_codes=True)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)
data = pd.read_csv("../data/raw/FIFA_players_21.csv")
```

# INITIAL ANALYSIS
Basic functions

```python
print('Shape of the data:',data.shape)
print("\n****************************\n")
print('Columns of the data:',data.columns)
print("\n****************************\n")
print('Number of unique values:\n',data.nunique())
print("\n****************************\n")
print('\nNumber of duplicated values:',data.duplicated().sum())
data.describe()
```

Verify data types

```python
for col in data.columns:
    print(f"{col}: {data[col].dtype}")
```

Copy dataset

```python
df = data.copy()
df.sample(3)
```

Drop useless columns for value prediction because these columns show ingame performance in each position and chemistry boost based on where you place the player

```python
df.drop(["ls","st","rb","rcb","cb","lcb","lb","rwb","rdm","cdm","ldm","lwb","rm","rcm","cm","rs","lcm","lm","ram","cam","lam","rw","rf","cf","lf","lw"],axis=1,inplace=True)
```

# FEATURE ENGINEERING

Convert object features to datetime

```python
df["dob"]=pd.to_datetime(df["dob"])
df["joined"]=pd.to_datetime(df["joined"])
for col in df.columns:
    print(f"{col}: {df[col].dtype}")
```

Body type conversion

```python
def simplify_body_type(body):
    body = str(body).strip()

    if body in ['Messi', 'C. Ronaldo', 'Neymar', 'Mohamed Salah', 'Shaqiri', 'Akinfenwa', 'Courtois']:
        return 'Unique'
    elif 'PLAYER_BODY_TYPE_' in body:
        return 'Unique'
    elif body == 'Lean':
        return 'Lean'
    elif body == 'Normal':
        return 'Average'
    elif body == 'Stocky':
        return 'Stocky'
    else:
        return 'Unknown'

df['body_type'] = df['body_type'].apply(simplify_body_type)
print(df['body_type'].value_counts())
```

Transforming player_traits column to show the amount of traits

```python
df['player_traits'] = df['player_traits'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)
df.head(5)
```

Converting value features to millions

```python
df["value_eur"]=df["value_eur"]/1000000
df["wage_eur"]=(df["wage_eur"]*52)/1000000 #yearly wage
df["release_clause_eur"]=df["release_clause_eur"]/1000000

df.rename(columns={"value_eur":"value_eur_m","wage_eur":"wage_eur_year_m","release_clause_eur":"release_clause_eur_m"},inplace=True)
```

Creating a column to show who is a goalkeeper and who isnt

```python
df['is_goalkeeper'] = df['player_positions'].str.contains('GK', na=False).astype(int)
```

Simplify team_position column

```python
print(df['team_position'].unique())
df["team_position"]=df["team_position"].apply(lambda x: "attack" if x in ["LS","ST","RS","LW","LF","CF","RF","RW"] 
                                              else ("defense" if x in ["LWB","LB","LCB","CB","RCB","RB"] 
                                                else ("midfield" if x in ["LDM","CDM","RDM","LM","LCM","CM","RCM","RM","LAM","CAM","RAM"] 
                                                     else("goalkeeper" if x in ["GK"] else "substitute"))))
```

Show player versatility by how many positions they can play

```python
df['versatility'] = df['player_positions'].str.count(',') + 1
```

# MISSING VALUES

```python
missing_percent = data.isnull().mean() * 100
print(missing_percent.sort_values(ascending=False))
```

Deleting column with less than 90% of values (defending_marking, loaned_from, nation_jersey_number, nation_position, player_tags)

```python
for i in df.columns:
    if df[i].isnull().sum()>len(df)*0.90:
        df.drop(i,axis=1,inplace=True)
```

Removing players that have value_eur and wage_eur = 0

```python
df=df[(df['value_eur_m'] > 0) & (df['wage_eur_year_m'] > 0)].copy()
print('Shape of the df:',df.shape)
```

Fill null release_clause with avg ratio between value and release clause

```python
mask = (df['value_eur_m'].notnull()) & \
       (df['release_clause_eur_m'].notnull()) & \
       (df['value_eur_m'] > 0)


average_ratio = (df.loc[mask, 'release_clause_eur_m'] / df.loc[mask, 'value_eur_m']).mean()
print(f"Average release clause to value ratio: {average_ratio:.2f}x")

df['release_clause_eur_m'] = df['release_clause_eur_m'].fillna(df['value_eur_m'] * average_ratio)

df['release_clause_eur_m'] = df['release_clause_eur_m'].round(1)
```

Fill field players and goal-keepers ratings that dont concern them with 0

```python
need_fill = df.select_dtypes(include=['int64', 'float64']).columns
df[need_fill] = df[need_fill].fillna(0)
```

add placeholder dates for joined column where players were loaned to another club

```python
df.loc[df['joined'].isna(), 'joined'] = pd.to_datetime(
    (df.loc[df['joined'].isna(), 'contract_valid_until'] - 4).astype(int).astype(str) + '-01-01',
    format='%Y-%m-%d'
)
df.isnull().sum()
```

# OUTLIER DETECTION

```python
num=df.select_dtypes(include=['int64','float64'])
num.drop(["sofifa_id"],axis=1,inplace=True)

print()
fig, axs = plt.subplots(nrows=9, ncols=3, figsize=(18, 25))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("Outlier Detection (Filtered by Position)", fontsize=20, y=1.02)

for idx, (col, ax) in enumerate(zip(num.iloc[:, 0:27], axs.ravel())):
    if col in gk_stats:
        sns.boxplot(x=num.loc[df['is_goalkeeper'] == 1, col], ax=ax)
        ax.set_title(f"{col.upper()} (Goalkeepers Only)", fontsize=12)
    elif col in field_stats:
        sns.boxplot(x=num.loc[df['is_goalkeeper'] == 0, col], ax=ax)
        ax.set_title(f"{col.upper()} (Field Players Only)", fontsize=12)
    else:
        sns.boxplot(x=num[col], ax=ax)
        ax.set_title(col.upper(), fontsize=12)
    
    ax.set_xlabel("")

plt.tight_layout()
save_path = os.path.join(output_dir_plot, "Outlier Detection.png")
plt.savefig(save_path)
plt.show()
plt.close()
gk_zero_diving = df[(df['is_goalkeeper'] == 1) & (df['gk_diving'] == 0)]
print(gk_zero_diving)
```

Taking care of said outliers

```python
import pandas as pd
def outlier_limits(dataframe,col_name,q1=0.25,q3=0.75):    
    Q1 = dataframe[col_name].quantile(q1) 
    Q3 = dataframe[col_name].quantile(q3)  
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return lower_bound,upper_bound

def check_outliers(dataframe,col_name):
    lower_bound,upper_bound=outlier_limits(dataframe,col_name)
    outliers = dataframe[(dataframe[col_name] < lower_bound) | (dataframe[col_name] > upper_bound)]
    print(f"Found {len(outliers)} outliers in {col_name}:")

def replace_with_limits(dataframe, col):
    low_limit, up_limit = outlier_limits(dataframe, col)
    if low_limit > 0:
        dataframe.loc[(dataframe[col] < low_limit), col] = low_limit
        dataframe.loc[(dataframe[col] > up_limit), col] = up_limit
    else:
        dataframe.loc[(dataframe[col] > up_limit), col] = up_limit

print("Age before: ")
check_outliers(df,"age")
replace_with_limits(df,"age")
print("Age after: ")
check_outliers(df,"age")

print("Overall before: ")
check_outliers(df,"overall")
```
