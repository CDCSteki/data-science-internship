# DATA CLEANING
Create a copy of the DataFrame

```python
df = data.copy()
df.sample(3)
```

Deleting column with less than 90% of values (defending_marking, loaned_from, nation_jersey_number, nation_position, player_tags)

```python
for i in df.columns:
    if df[i].isnull().sum()>len(df)*0.90:
        df.drop(i,axis=1,inplace=True)
df.isnull().sum()
```

Convert object features to datetime

```python
df["dob"]=pd.to_datetime(df["dob"])
df["joined"]=pd.to_datetime(df["joined"])
for col in df.columns:
    print(f"{col}: {df[col].dtype}")
```

Body type conversion

```python
def simplify_body_type(body):
    body = str(body).strip()

    if body in ['Messi', 'C. Ronaldo', 'Neymar', 'Mohamed Salah', 'Shaqiri', 'Akinfenwa', 'Courtois']:
        return 'Unique'
    elif 'PLAYER_BODY_TYPE_' in body:
        return 'Unique'
    elif body == 'Lean':
        return 'Lean'
    elif body == 'Normal':
        return 'Average'
    elif body == 'Stocky':
        return 'Stocky'
    else:
        return 'Unknown'

df['body_type'] = df['body_type'].apply(simplify_body_type)
print(df['body_type'].value_counts())
```

Transforming player_traits column to show the amount of traits

```python
df['player_traits'] = df['player_traits'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)
df.head(5)
```

Removing players that have value_eur and wage_eur = 0

```python
df=df[(df['value_eur'] > 0) & (df['wage_eur'] > 0)].copy()
print('Shape of the df:',df.shape)
df.isnull().sum()
```

Fill release_clause with 0 for players that are loaned because they dont have a release clause at the borrowing club

```python
df['release_clause_eur'] = df['release_clause_eur'].fillna(0)
df.isnull().sum()
```
