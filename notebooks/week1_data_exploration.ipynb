# DATA CLEANING
Create a copy of the DataFrame

```python
df = data.copy()
df.sample(3)
```

Deleting column with less than 90% of values (defending_marking, loaned_from, nation_jersey_number, nation_position, player_tags)

```python
for i in df.columns:
    if df[i].isnull().sum()>len(df)*0.90:
        df.drop(i,axis=1,inplace=True)
```

Convert object features to datetime

```python
df["dob"]=pd.to_datetime(df["dob"])
df["joined"]=pd.to_datetime(df["joined"])
for col in df.columns:
    print(f"{col}: {df[col].dtype}")
```

Body type conversion

```python
def simplify_body_type(body):
    body = str(body).strip()

    if body in ['Messi', 'C. Ronaldo', 'Neymar', 'Mohamed Salah', 'Shaqiri', 'Akinfenwa', 'Courtois']:
        return 'Unique'
    elif 'PLAYER_BODY_TYPE_' in body:
        return 'Unique'
    elif body == 'Lean':
        return 'Lean'
    elif body == 'Normal':
        return 'Average'
    elif body == 'Stocky':
        return 'Stocky'
    else:
        return 'Unknown'

df['body_type'] = df['body_type'].apply(simplify_body_type)
print(df['body_type'].value_counts())
```

Transforming player_traits column to show the amount of traits

```python
df['player_traits'] = df['player_traits'].fillna('').apply(lambda x: len(x.split(',')) if x else 0)
df.head(5)
```

Removing players that have value_eur and wage_eur = 0

```python
df=df[(df['value_eur'] > 0) & (df['wage_eur'] > 0)].copy()
print('Shape of the df:',df.shape)
```

Fill release_clause with 0 for players that are loaned because they dont have a release clause at the borrowing club

```python
df['release_clause_eur'] = df['release_clause_eur'].fillna(0)
```

Fill field players and goal-keepers rating that dont concern them with 0

```python
need_fill = df.select_dtypes(include=['int64', 'float64']).columns
df[need_fill] = df[need_fill].fillna(0)
```

add placeholder dates for joined column where players were loaned to another club

```python
df.loc[df['joined'].isna(), 'joined'] = pd.to_datetime(
    (df.loc[df['joined'].isna(), 'contract_valid_until'] - 4).astype(int).astype(str) + '-01-01',
    format='%Y-%m-%d'
)
df.isnull().sum()
```

Looking for outliers

```python
num=df.select_dtypes(include=['int64','float64'])
num.drop(["sofifa_id"],axis=1,inplace=True)


fig, axs = plt.subplots(nrows=9, ncols=3, figsize=(15, 12))
plt.subplots_adjust(hspace=2)
fig.suptitle("Outlier Detection in selected significant numerical features", fontsize=18, y=0.95)

for ticker, ax in zip(num.iloc[:,0:27], axs.ravel()):
    sns.boxplot(data=num, x=ticker, ax=ax)
    ax.set_title(ticker.upper())

plt.show()
```

Taking care of said outliers

```python
import pandas as pd
def outlier_limits(dataframe,col_name,q1=0.25,q3=0.75):    
    Q1 = dataframe[col_name].quantile(q1) 
    Q3 = dataframe[col_name].quantile(q3)  
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return lower_bound,upper_bound

def check_outliers(dataframe,col_name):
    lower_bound,upper_bound=outlier_limits(dataframe,col_name)
    outliers = dataframe[(dataframe[col_name] < lower_bound) | (dataframe[col_name] > upper_bound)]
    print(f"Found {len(outliers)} outliers in {col_name}:")

def replace_with_limits(dataframe, col):
    low_limit, up_limit = outlier_limits(dataframe, col)
    if low_limit > 0:
        dataframe.loc[(dataframe[col] < low_limit), col] = low_limit
        dataframe.loc[(dataframe[col] > up_limit), col] = up_limit
    else:
        dataframe.loc[(dataframe[col] > up_limit), col] = up_limit

print("Age before: ")
check_outliers(df,"age")
replace_with_limits(df,"age")
print("Age after: ")
check_outliers(df,"age")

print("Wage before: ")
check_outliers(df,"wage_eur")
replace_with_limits(df,"wage_eur")
print("Wage after: ")
check_outliers(df,"wage_eur")
```
